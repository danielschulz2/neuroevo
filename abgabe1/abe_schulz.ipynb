```python

def exercise2(length, flip_prob):
    # Liste erstellen zufällig mit 0 oder 1 
    bit_list = [random.choice([0, 1]) for _ in range(length)]

    #! Gut, ihr könnt alternativ auch all(bit_list) verwenden, da das die boolische Wahrheit aller Elemente der Liste
    #! überprüft.
    # while any(bit == 0 for bit in bit_list):  # Solange noch nicht alle Bits auf 1
    while not all(bit_list):  # Solange noch nicht alle Bits auf 1
        for i in range(len(bit_list)):
            if random.random() < flip_prob:  # mit angegebener Wkeit Bitflip durchführen
                bit_list[i] ^= 1  # XOR durchführen

    #! Alternativ dazu, die Bits einzeln zu flippen, könnt ihr auch "Maske" zum Flippen der Bits erstellen.
    #! Dann könnt ihr die gesamte Liste mit einem XOR-Operator auf einmal flippen. Das könnte so aussehen:
    #! flip = np.random.random((len(bit_list))) < 0.1
    #! bit_list ^= flip
    #! Das geht insgesamt noch etwas schneller

    return bit_list
    
test.test_task_1_2(exercise2, {"length": 5, "flip_prob": 0.1})
```

## Aufgabe 3

```python
def run_bitflip_simulation():
    # Liste der Länge 5 mit zufällig 0 oder 1 befüllt 
    bit_list = [random.choice([0, 1]) for _ in range(5)]
    
    # Counter für die Anzahl der Bitflips
    flip_count = 0
    
    # alle Werte müssen auf 1
    #! Auc hhier funktioniert der Trick mit all() wieder.
    while any(bit == 0 for bit in bit_list):
        for i in range(len(bit_list)):
            if random.random() < 0.1:  # Mit 10% Wahrscheinlichkeit Bitflip 
                bit_list[i] ^= 1  # XOR-Operator für Bitflip
        #! Hier wird nicht die Anzahl der Bitflips gezählt, sondern die Anzahl der Iterationen.
        #! Um die Anzahl der Bitflips zu zählen, muss der Aufruf immer mit dem Aufruf des XOR-Operators
        #! erfolgen.
        flip_count += 1  # Zähle den Bitflip hoch
    
    return flip_count

# Liste für die Anzahl der Bitflips für 100 Ausführungen
flip_counts = []

# 100 mal ausführen und Daten sammeln
for _ in range(100):
    flip_counts.append(run_bitflip_simulation())

# Visualisierung für die Anzahl der Bitflips
plt.hist(flip_counts, bins=range(min(flip_counts), max(flip_counts) + 2), edgecolor='black')
plt.title('Verteilung der Anzahl der Bitflips (100 Ausführungen)')
plt.xlabel('Anzahl der Bitflips')
plt.ylabel('Häufigkeit')
plt.show()
```

```python
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

def rosenbrock_plot():
    # Grid vorbereiten
    plotx = np.linspace(-2, 2, 1000)
    ploty = np.linspace(-1, 3, 1000)
    X, Y = np.meshgrid(plotx, ploty)
    Z = rosenbrock(X,Y)
    
    # Plotten
    plt.figure(figsize=(8,6))
    #! Schön, dass ihr hier LogNorm() verwendet!
    plt.contourf(X, Y, Z, norm=LogNorm())
    plt.plot(1, 1, 'ro', markersize=3)
    plt.colorbar()
    plt.title("Rosenbrock Funktion")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

# initiale Population erzeugen
def init_population(n=10):
    pop = []

    #! Die Erstellung einer Population mit der Größe n lässt sich mit Numpy auch noch beschleunigen.
    #! Das könnte dann z.B. so aussehen.
    #! x = np.random.uniform(-2.0, 2.0, n)
    #! y = np.random.uniform(-1.0, 3.0, n)
    #! pop = np.column_stack((x, y))
    #! Der Effekt wird dabei jedoch erst bei größeren Populationen spürbar. Für n=10 macht es keinen
    #! relevanten Unterschied.
    for _ in range(n):
        x = random.uniform(-2.0, 2.0)
        y = random.uniform(-1.0, 3.0)
        pop.append((x, y))
    return pop

# Fitness
def fitness(ind):
    return rosenbrock(ind[0], ind[1])

def show_best_three():
    # Besten 3 Kandidaten
    population = init_population(10)
    population.sort(key=fitness)
    best3 = population[:3]
    
    # Plotten
    #! An dieser Stelle hättet ihr auch die rosenbrock_plot() Funktion nutzen können
    #! um Code-Wiederholungen zu vermeiden.
    plotx = np.linspace(-2, 2, 1000)
    ploty = np.linspace(-1, 3, 1000)
    X, Y = np.meshgrid(plotx, ploty)
    Z = rosenbrock(X,Y)
    plt.figure(figsize=(8,6))
    plt.contourf(X, Y, Z, norm=LogNorm())
    for p in best3:
        plt.plot(p[0], p[1], 'ro',markersize=5)
    plt.colorbar()
    plt.title("Beste 3 Lösungen (Initial)")
    plt.show()
    
rosenbrock_plot()

show_best_three()
```

```python
#! Diesen Code habt ihr auch wiederholt:
#! =========================================
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# initiale Population erzeugen
def init_population(n=10):
    pop = []
    for _ in range(n):
        x = random.uniform(-2.0, 2.0)
        y = random.uniform(-1.0, 3.0)
        pop.append((x, y))
    return pop

# Fitness
def fitness(ind):
    return rosenbrock(ind[0], ind[1])
#! =========================================

#! Das ihr hier die Hyperparameter des EAs in den Funktionsparametern übergebt ist gut.
#! Achtet darauf, dass ihr alle Hyperparameter einheitlich an einer Stelle sammelt (also z.B. auch
#! die Anzahl an Generationen). Alternativ zu den Funktionsparametern könnt ihr die Hyperparameter
#! auch einmalig an einer Stelle im Code als Konstanten festlegen.
def evolutionary_rosenbrock_algorithm(pop_size = 100, select_size = 6, show_plot=False):
    #Constraint in der Aufgabenstellung maximal 100 gens
    #! Konstanten sollten in Python immer außerhalb von Funktionen deklariert werden.
    GENS = 100

    #! Auch hier könnt ihr die rosenbrock_plot() Funktion nutzen, um den Code nicht zu wiederholen.
    #Darstellung der Funktion
    plotx = np.linspace(-2, 2, 1000)
    ploty = np.linspace(-1, 3, 1000)
    X, Y = np.meshgrid(plotx, ploty)
    Z = rosenbrock(X,Y)

    #? Wieso verwendet ihr hier nicht init_population()?
    # initiale Population mit zufälligen Werten aus dem Wertebereich
    population = []
    for _ in range(pop_size):
        rx = np.random.uniform(-2, 2 + np.finfo(float).eps)
        ry = np.random.uniform(-1, 3 + np.finfo(float).eps)
        population.append((rx ,ry))

    for i in range(GENS):
        #Top 6 von population sortiert nach rosenbrock
        population.sort(key=lambda p: rosenbrock(p[0], p[1]))
        bests = population[:select_size]

        #! Auch hier könnt ihr die rosenbrock_plot() Funktion nutzen, um den Code nicht zu wiederholen.
        #if Abfrage eingebaut, da es nervig beim Testing war
        if show_plot:
            clear_output(wait=True)
            plt.figure(figsize=(8,6))
            plt.contourf(X, Y, Z, norm=LogNorm())
            plt.colorbar()
            plt.xlabel('x')
            plt.ylabel('y') 
            
            x_points = [b[0] for b in bests]
            y_points = [b[1] for b in bests]
            
            plt.plot(x_points, y_points, 'ro', markersize=5)
            plt.title("Generation " + str(i))
            plt.show()
            time.sleep(0.1)
            plt.close()
            
        
        #Abbruchbedingung: nicht unnötig 100 Generationen durchgehen
        if rosenbrock(bests[0][0], bests[0][1]) <= 1e-04:
            return bests[0][0],bests[0][1]

        children = []
        for _ in range(pop_size):
            # zufällige Parents aus den besten Punkten
            parent1 = random.choice(bests)
            parent2 = random.choice(bests)

            #! Ihr verbindet hier das Crossover mit der Mutation und führt diese beiden Schritte auf einmal aus.
            #! Das ist technisch nicht falsch, jedoch reduziert es die Lesbarkeit des EAs und erschwert
            #! das Debugging. Es verhindert außerdem z.B. nur einige Kinder nach einer Mutation-Rate zu
            #! mutieren und andere nicht.
            #! Ich würde euch daher für zukünftige Aufgaben empfehlen die Schritte aus der EA-Schleife auch im Code
            #! voneinander zu trennen und am besten in separate Funktionen auszulagern.
            #Wir nehmen den Mittelwert zwei von den besten Punkten und machen eine zusätzliche Mutation
            cx = (parent1[0] + parent2[0]) / 2 + np.random.uniform(-0.1,0.1)
            cy = (parent1[1] + parent2[1]) / 2 + np.random.uniform(-0.1,0.1)

            children.append((cx,cy))

        #! An dieser Stelle ersetzt ihr die gesamte Population durch die Kinder und verwerft damit die
        #! fittesten Individuen aus der Elterngeneration (bests). Das ist eine mögliche Strategie,
        #! sie hat jedoch zur Folge, dass euer EA vor allem in der Rosenbrock-Funktion Rückschritte
        #! macht, weil er die fittesten Individuen verwirft und durch möglicherweise weniger fitte Kinder ersetzt.
        population = children

    #wenn nach 100 gens Abbbruchbedingung nicht erfüllt ist, dann besten (minimalen) Wert returnen 
    population.sort(key=lambda p: rosenbrock(p[0], p[1]))
    print(population[0][0], population[0][1])
    return population[0][0], population[0][1]

# evolutionary_rosenbrock_algorithm(100, 6, True)
# test.test_task_1_4(evolutionary_rosenbrock_algorithm, {"pop_size": 100, "select_size":6, "show_plot": False})

# eigene Testmethode um zu sehen, ob Erfolgsquote von über 70% eingehalten ist
def test_ea():
    inrange_counter = 0
    
    for i in range(100):
        my_result = evolutionary_rosenbrock_algorithm()
        value_of_rosenbrock = rosenbrock(my_result[0], my_result[1])
        if value_of_rosenbrock < 1e-4:
            inrange_counter += 1
            
    print("Erfolgsquote:", inrange_counter / 100)

test_ea()
```
